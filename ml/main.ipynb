```python
import os
import pandas as pd
import re
from collections import Counter
import matplotlib.pyplot as plt
```

```python
BASE_DIR = os.getcwd()
DIRECTORY_PATH = os.path.dirname(BASE_DIR)
df_1 = pd.read_csv(DIRECTORY_PATH + '\\parser\\fake_reviews.csv')
df_2 = pd.read_csv(DIRECTORY_PATH + '\\parser\\fake_feedbacks.csv')
```

```python
df = pd.concat([df_1, df_2], ignore_index=True, axis=0)
```

```python
df
```

```python
all_text = ' '.join(df['User review'])
```

```python
words = re.findall(r'\b\w{4,}\b', all_text.lower())
```

```python
word_counts = Counter(words)
```

```python
most_common_words = word_counts.most_common(20)
```

```python
most_common_words
```

```python
most_words = [word for word, freq in most_common_words]
```

```python
frequencies = [freq for word, freq in most_common_words] 
```

```python
plt.figure(figsize=(28, 15))
plt.xlabel('Слово')
plt.ylabel('Частота')
plt.plot(most_words, frequencies)
plt.show()
```

```python
df_1 = pd.read_csv(DIRECTORY_PATH + '\\parser\\dataframe.csv')
df_2 = pd.read_csv(DIRECTORY_PATH + '\\parser\\feedbacks.csv')
```

```python
df = pd.concat([df_1, df_2], ignore_index=True, axis=0)
```

```python
all_text = ' '.join(df['User review'])
words = re.findall(r'\b\w{4,}\b', all_text.lower())
word_counts = Counter(words)
most_common_words = word_counts.most_common(20)
most_words = [word for word, freq in most_common_words]
frequencies = [freq for word, freq in most_common_words] 
```
