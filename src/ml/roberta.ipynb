```python
import json
import os
import pandas as pd
from transformers import RobertaTokenizer, TFRobertaModel
```

```python
BASE_DIR = os.getcwd()
DIRECTORY_PATH = os.path.dirname(BASE_DIR)
DF_PATH = DIRECTORY_PATH + "\\parser\\dataframes\\preprocessed_main_df.csv"
```

```python
df = pd.read_csv(DF_PATH)
df = df.copy()
df['User review'] = df['User review'].apply(json.loads)
```

```python
tokenizer = RobertaTokenizer.from_pretrained("roberta-base")
model = TFRobertaModel.from_pretrained("roberta-base")
```

```python
reviews = df['User review']
```

```python
input_ids = [tokenizer.convert_tokens_to_ids(tokens) for tokens in reviews]
```

```python
df['User review'] = input_ids
```
